```python
with open('poems.txt', 'rb') as f:
    content = f.read().decode(errors='ignore')
    content = content.replace(',', ' ').replace('.', ' ').replace('?', ' ').replace('!', ' ').replace(':', ' ').replace(';', ' ')
    words = content.split()
f.close()

word_counter = {}
for word in words:
    word_counter[word] = word_counter.get(word, 0) + 1

for word, count in sorted(word_counter.items(), key=lambda x: x[1], reverse=True):
    print(f'{word}: {count}')   
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.tag import pos_tag
from nltk.probability import FreqDist
import string


nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('stopwords')

with open('poems.txt', 'rb') as f:
    content = f.read().decode(errors='ignore')

words = word_tokenize(content.lower())
tagged_words = pos_tag(words)
additional_removes = ['”', '“', "’", 
                      "you", "your", "note", "s", 
                      "i", "it", "A", "be", 
                      "was", "were", "not",
                      "our", "they", "their", "we",
                      "will", "how", "can", "been",
                      "what", "had", "why", "who"]

filtered_words = [
                word for word, pos
                in tagged_words 
                if pos not in ['DT', 'IN', 'TO', 'CC', 'VBZ', 'VBP'] 
                and word not in string.punctuation
                and word not in additional_removes
                ]

word_counter = FreqDist(filtered_words)

for word, count in word_counter.most_common():
    print(f'{word}: {count}')
```
